```python
import pandas as pd
import bambi as bmb
import pingouin as pg
import joblib
from os import listdir
from os.path import join
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
import arviz as az

from fooof import FOOOFGroup

from neurodsp.plts.spectral import plot_power_spectra

INDIR = '/mnt/obob/staff/fschmidt/cardiac_1_f/data/ei_resting'
all_files = listdir(INDIR)
```

```python
#ecg = [(joblib.load(join(INDIR, file))['ecg_scores'] > 0.5).sum() > 0 for file in all_files]
```

```python
all_df, all_psd = [], []
for file in all_files:
    cur_data = joblib.load(join(INDIR, file))
    
    if 'ecg_scores' in cur_data.keys() and (cur_data['ecg_scores'] > 0.5).sum() > 0:
    
        all_psd.append(pd.DataFrame({'ECG Electrode': cur_data['data_heart']['ecg']['psd'][0,:],
                                     'Heart Component Magnetometers': cur_data['data_heart']['mag']['psd'].mean(axis=0),
                                     'Magnetometers': cur_data['data_brain']['mag']['psd'].mean(axis=0),
                                     'Frequency (Hz)': cur_data['data_heart']['ecg']['freqs'],
                                     'subject_id': cur_data['subject_id']}))
    
    
        all_df.append(pd.DataFrame({'heart_slope_mag': cur_data['data_heart']['mag']['exponents'],
                                'brain_slope': cur_data['data_brain']['mag']['exponents'],
                                'brain_no_ica': cur_data['data_no_ica']['mag']['exponents'],
                                'heart_slope': cur_data['data_heart']['ecg']['exponents'][0],
                                'heart_offset': cur_data['data_heart']['ecg']['offsets'][0],
                                'r2_heart': cur_data['data_heart']['ecg']['fg'].get_params('r_squared')[0],
                                'r2_brain': cur_data['data_brain']['mag']['fg'].get_params('r_squared'),
                                'r2_no_ica': cur_data['data_no_ica']['mag']['fg'].get_params('r_squared'),
                                'r2_heart_mag': cur_data['data_heart']['mag']['fg'].get_params('r_squared'),
                                'heart_slope_avg': cur_data['data_heart']['mag']['aps_mag']['Exponent'].mean(),
                                'brain_slope_avg': cur_data['data_brain']['mag']['aps_mag']['Exponent'].mean(),
                                'no_ica_slope_avg': cur_data['data_no_ica']['mag']['aps_mag']['Exponent'].mean(),
                                'heart_offset_avg': cur_data['data_heart']['mag']['aps_mag']['Offset'].mean(),
                                'brain_offset_avg': cur_data['data_brain']['mag']['aps_mag']['Offset'].mean(),
                                'no_ica_offset_avg': cur_data['data_no_ica']['mag']['aps_mag']['Offset'].mean(),
                                'channel': np.arange(102),
                                'subject_id': cur_data['subject_id'],
                                'age': cur_data['age'],
                                'meas_date': cur_data['meas_date'],
                                'days': cur_data['days']}))
df_cmb = pd.concat(all_df)
df_cmb_psd = pd.concat(all_psd)
```

```python
subject_list = df_cmb_psd['subject_id'].unique()
```

```python
all_r_heart, all_r_brain = [], []


for subject in subject_list:
    cur_df = df_cmb_psd.query(f'subject_id == "{subject}"')

    all_r_brain.append(pg.corr(cur_df['Magnetometers'], cur_df['ECG Electrode'])['r'])
    all_r_heart.append(pg.corr(cur_df['Heart Component Magnetometers'], cur_df['ECG Electrode'])['r'])
#plt.plot(cur_df['Frequency (Hz)'], cur_df['Heart Component Magnetometers'])
#plt.xscale('log')
#plt.yscale('log')

```

```python
heart_heart_stats, heart_stats, brain_stats, no_ica = [], [], [], []

for channel in np.arange(102):
    cur_df = df_cmb.query(f'channel == {channel}')
    
    heart_stats.append(pg.corr(cur_df['age'], cur_df['heart_slope_mag'], method='shepherd'))
    brain_stats.append(pg.corr(cur_df['age'], cur_df['brain_slope'], method='shepherd'))
    no_ica.append(pg.corr(cur_df['age'], cur_df['brain_no_ica'], method='shepherd'))
```
